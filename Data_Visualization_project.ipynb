BIVARIATE ANALYSIS

```python
cross_tab = pd.crosstab(data['Gender'], data['Smoking'])
cross_tab.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Smoking by Gender')
plt.show()
```

```python
import matplotlib.pyplot as plt
gender_counts = data.groupby(['Air Pollution', 'Gender']).size().unstack()

gender_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm')

plt.xlabel('Air Pollution Level')
plt.ylabel('Count')
plt.title('Composition of Gender by Air Pollution Level')
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.scatter(data['Age'], data['Air Pollution'])
plt.xlabel('Age')
plt.ylabel('Air Pollution')
plt.title('Scatter Plot: Age vs Air Pollution')

plt.subplot(2, 2, 2)
sns.boxplot(data=data, x='Gender', y='Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot: Age by Gender')

plt.subplot(2, 2, 3)
sns.swarmplot(x=data['Smoking'], y=data['Air Pollution'], data=data)
plt.xlabel('Smoking')
plt.ylabel('Air Pollution')
plt.title('Swarm Plot: Smoking vs Air Pollution')

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='Smoking', y='Age',data=data)
plt.xlabel('Smoking')
plt.ylabel('Age')
plt.title('Box Plot: Age by Smoking Status')
plt.show()
```

```python
sns.jointplot(x='Age', y='Genetic Risk', data=data, kind='hex', color='purple')
plt.xlabel('Age')
plt.ylabel('Genetic Risk')
plt.suptitle('Joint Plot: Age vs. Genetic Risk')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='Gender', y='Age', data=data, palette='pastel', inner='quart')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender (Violin Plot)')
plt.show()
```

MULTIVARIATE ANALYSIS

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap')

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
level_counts = data['Level'].value_counts()
plt.pie(level_counts, labels=level_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Level Distribution')
plt.axis('equal')
plt.show()
```

```python
selected_column = 'Gender'

value_counts = data[selected_column].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
plt.title(f'Distribution of {selected_column}')
plt.axis('equal')

plt.show()
```

```python
sns.pairplot(data=data, vars=['Age', 'Air Pollution', 'Alcohol use'])
plt.show()
```

```python
sns.set(style="ticks")
numeric_vars = ['Age', 'Air Pollution', 'Obesity', 'Genetic Risk']
sns.pairplot(data, vars=numeric_vars, hue='Level', palette='viridis', markers=["o", "s", "D"])
plt.suptitle('Scatter Matrix of Numeric Variables by Level')
plt.show()
```
