```python
# Clean and convert the 'Gears' column to numeric (assuming it contains non-numeric data)
df['Gears'] = pd.to_numeric(df['Gears'], errors='coerce')

# Create a frequency (count) plot for the 'Gears' column
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Create a custom color palette for the bars
colors = sns.color_palette("viridis", len(df['Gears'].unique()))

# Create the count plot with custom colors
sns.countplot(data=df, x='Gears', palette=colors)

# Adding a gradient background
ax = plt.gca()
ax.set_facecolor('#f0f0f0')

# Customizing the title and labels
plt.title('Frequency Plot of Gears', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Number of Gears', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Adding data labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height() + 50),
                ha='center', va='center', fontsize=12, color='black')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adding a color bar legend
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=min(df['Gears']), vmax=max(df['Gears'])))
sm.set_array([])  # You need to set an empty array to create the ScalarMappable
cbar = plt.colorbar(sm, pad=0.1)
cbar.set_label('Count', fontsize=12)

# Adding a decorative grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python


# Extract unique values and their counts in the 'Emission_Norm' column
emission_norm_counts = df['Emission_Norm'].value_counts().reset_index()
emission_norm_counts.columns = ['Category', 'Count']

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))
sns.set_style("whitegrid")  # Set the background style

# Bar plot for Emission Norm distribution
sns.barplot(data=emission_norm_counts, x='Category', y='Count', palette='viridis', ax=ax1)
ax1.set_xlabel('Emission Norm', fontsize=14)
ax1.set_ylabel('Count', fontsize=14)
ax1.set_title('Distribution of Emission Norm', fontsize=16)
ax1.tick_params(axis='x', labelrotation=45, labelsize=12)

# Create a secondary y-axis for the line plot
ax2 = ax1.twinx()

# Line plot for cumulative count
cumulative_counts = emission_norm_counts['Count'].cumsum()
sns.lineplot(x=emission_norm_counts.index, y=cumulative_counts, marker='o', color='blue', ax=ax2)
ax2.set_ylabel('Cumulative Count', fontsize=14)

# Adjust the x-axis labels to match the categories
ax2.set_xticks(range(len(emission_norm_counts)))
ax2.set_xticklabels(emission_norm_counts['Category'], rotation=45, fontsize=12)

# Add a legend for clarity
ax1.legend(['Count'], loc='upper left')
ax2.legend(['Cumulative Count'], loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV data into a DataFrame
df = pd.read_csv("/content/drive/MyDrive/Data visualization/sid dv car.csv", encoding='ISO-8859-1')

# Create a pie chart for the 'Doors' column
plt.figure(figsize=(8, 8))

# Count the occurrences of each unique value in 'Doors'
door_counts = df['Doors'].value_counts()

# Create labels for the pie chart
labels = door_counts.index

# Create values for the pie chart
sizes = door_counts.values

# Define color options for the pie chart
colors = ['red', 'yellow', 'blue', 'pink']

# Define hatch patterns for the pie chart
hatches = ['**o', '\\', '.||.', 'o.o']

# Create the pie chart with hatch patterns
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    pctdistance=1.3,  # Set the percentage label's distance from the center
)

# Add hatch patterns to the slices
for i, patch in enumerate(plt.gca().patches):
    patch.set_hatch(hatches[i])

# Set the title
plt.title('Distribution of Doors', fontsize=16)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV data into a DataFrame
df = pd.read_csv("/content/drive/MyDrive/Data visualization/sid dv car.csv", encoding='ISO-8859-1')

# Clean and convert the 'Ex-Showroom_Price' column to float
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)

# Calculate the average Ex-Showroom Price for each car make
average_prices_by_make = df.groupby('Make')['Ex-Showroom_Price'].mean().reset_index()

# Sort the data by average Ex-Showroom Price in descending order
sorted_data = average_prices_by_make.sort_values(by='Ex-Showroom_Price', ascending=False)

# Create a bar plot to visualize the most expensive car makes
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

sns.barplot(x='Make', y='Ex-Showroom_Price', data=sorted_data, palette='viridis')
plt.title('Average Ex-Showroom Price by Car Make', fontsize=16)
plt.xlabel('Make', fontsize=12)
plt.ylabel('Average Ex-Showroom Price', fontsize=12)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset from a CSV file with explicit data type specification
data_types = {
    "Make": str,  # Assuming Make is a string (categorical)
    "Ex-Showroom_Price": str  # Assuming Ex-Showroom_Price is initially read as a string
}

df = pd.read_csv("/content/drive/MyDrive/Data visualization/sid dv car.csv", encoding='ISO-8859-1', dtype=data_types)

# Clean and preprocess the "Ex-Showroom_Price" column
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)

# Get the top 10 unique values in the "Make" column
top_10_makes = df['Make'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 makes
df_filtered = df[df['Make'].isin(top_10_makes)]

# Create a box plot to visualize the relationship between Make and Ex-Showroom_Price
plt.figure(figsize=(10, 6))
sns.boxplot(x="Make", y="Ex-Showroom_Price", data=df_filtered)
plt.title("Bivariate Analysis: Top 10 Makes vs. Ex-Showroom Price")
plt.xlabel("Make")
plt.ylabel("Ex-Showroom Price")
plt.xticks(rotation=45)
plt.show()
```
